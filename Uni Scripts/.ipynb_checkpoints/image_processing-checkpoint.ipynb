```python
zdjecie = image.imread('zdjeciezwakacji.jpg')
plt.imshow(zdjecie)
```

```python
def zdjecie_wb_blurr(x):
    fotka = image.imread(x)
    z = ndimage.gaussian_filter(fotka, sigma=8)
    return plt.imshow(z, cmap=plt.cm.gray)
```

```python
zdjecie_wb_blurr('zdjeciezwakacji.jpg')
```

Zadanie 3

```python
def losuj_rnorm(x):
    r_norm = np.random.normal(2,10,x)
    plt.hist(r_norm, bins='auto')
    plt.title("Histogram rozkładu normalnego",fontdict={'color': 'pink'})
    return plt.show()
```

```python
losuj_rnorm(100)
```
